```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
```

```python
gdansk = pd.read_csv("Gdansk_dzielnice.csv")
gdynia = pd.read_csv("Gdynia_dzielnice.csv")
sopot = pd.read_csv("Sopot_dzielnice.csv")
```

```python
gdansk['Dzielnica'] = gdansk['Dzielnica_N']
gdansk.drop(columns=['Dzielnica_N', 'Dzielnica_Oryg'], inplace=True)
gdansk.drop(columns=['Unnamed: 0', 'Unnamed: 0.1'], inplace=True)
gdynia.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
sell_data = pd.concat([gdynia, gdansk, sopot], ignore_index=True)
```

```python
sell_data["Liczba łazienek"].unique()
```

```python
sell_data.loc[sell_data["Cena"].str[-1] == "€", ["Cena"]] = pd.to_numeric(sell_data[sell_data["Cena"].str[-1] == "€"]["Cena"].str.replace("€","").str.replace(" ", ""))*4.3
sell_data.loc[sell_data["Cena"].str[-3:] == "USD", ["Cena"]] = pd.to_numeric(sell_data[sell_data["Cena"].str[-3:] == "USD"]["Cena"].str.replace("USD","").str.replace(" ", ""))*4
```

```python
sell_data = sell_data.drop(sell_data[sell_data["Cena"]=="Zarezerwowano"].index)
sell_data["Cena"] = sell_data["Cena"].astype(str).str.replace("zł","").str.replace(" ", "")
sell_data["Cena"] = pd.to_numeric(sell_data["Cena"])
```

```python
sell_data["Miasto"] = sell_data["Miasto"].str.replace(",","")
```

```python
# pokoje sa
sell_data[sell_data["Miasto"] == "Sopot"]
```

```python
sell_data["Pokoje"].unique()
```

```python
sell_data["Pokoje"] = sell_data["Pokoje"].astype(str).str.replace("• ","")
sell_data["Pokoje"].unique()
```

```python
sell_data["Pokoje"] = sell_data["Pokoje"].astype(str).str.replace(" pokoje","")
sell_data["Pokoje"] = sell_data["Pokoje"].astype(str).str.replace(" pokoi","")
sell_data["Pokoje"].unique()
```

```python
sell_data[sell_data["Miasto"] == "Gdańsk"]
```

```python
sell_data["Pow. całkowita"] = sell_data["Pow. całkowita"].astype(str).str[:-3]
sell_data["Pow. całkowita"] = sell_data["Pow. całkowita"].astype(str).str.replace(",",".")
sell_data["Pow. całkowita"] = pd.to_numeric(sell_data["Pow. całkowita"])
```

```python
sell_data
```

```python
sell_data['Piętro'] = sell_data['Piętro'].fillna(sell_data['Piętro'].value_counts().index[0])
sell_data['Liczba pięter'] = sell_data['Liczba pięter'].fillna(sell_data['Liczba pięter'].value_counts().index[0])
sell_data['Piętro'].unique()
```

```python
sell_data.head()
```

```python
sell_data.info()
```

```python
sell_data.dropna(subset = ['Pokoje'], inplace=True)
sell_data.dropna(subset = ['Pow. całkowita'], inplace=True)
sell_data.dropna(subset = ['Dzielnica'], inplace=True)
```

```python
sell_data["Miasto"].unique()
```

```python
sell_data.loc[sell_data['Taras'] == 'Tak', 'Balkon'] = 'Tak'
sell_data.loc[sell_data['Pow. balkonu'].notna(), 'Balkon'] = 'Tak'
sell_data.loc[sell_data['Pow. loggii'].notna(), 'Balkon'] = 'Tak'
sell_data.loc[sell_data['Pow. tarasu'].notna(), 'Balkon'] = 'Tak'
sell_data.loc[sell_data['Pow. tarasu'].notna(), 'Balkon'] = 'Tak'

sell_data['Balkon'].fillna('Nie', inplace=True)
```

```python
data_cleaned = sell_data[['Cena', 'Miasto', 'Dzielnica', 'Pokoje', 'Pow. całkowita', 'Rynek', 'Rok budowy', 'Piętro', 'Liczba pięter', 'Balkon', 'Typ oferty']]
```

```python
data_cleaned["Cena/m2"] = round(data_cleaned["Cena"]/data_cleaned["Pow. całkowita"], 2)
```

```python
data_cleaned = data_cleaned.dropna()
```

```python
data_cleaned.info()
```

```python
data_cleaned["Miasto"].unique()
```

```python
data_cleaned['Pokoje'].unique()
```

```python
data_cleaned['Pokoje'] = data_cleaned['Pokoje'].str.strip()
data_cleaned['Pokoje'] = data_cleaned['Pokoje'].replace({'1 pokój': 1, '52 m²': 2, '45 m²': 2, '47 m²': 2})
data_cleaned['Pokoje'] = data_cleaned['Pokoje'].astype(int)
```

```python
data_cleaned['Rok budowy'].unique()
```

```python
data_cleaned = data_cleaned.drop(data_cleaned[data_cleaned['Rok budowy'] == '.'].index)
data_cleaned = data_cleaned.dropna()
data_cleaned['Rok budowy'] = data_cleaned['Rok budowy'].astype(int)
data_cleaned['Piętro'] = data_cleaned['Piętro'].astype(int)
data_cleaned['Liczba pięter'] = data_cleaned['Liczba pięter'].astype(int)
```

```python
data_cleaned["Miasto"].unique()
```

```python
# Gdynia - najem vs sprzedaz

data_cleaned[(data_cleaned['Cena'] > 100000) & (data_cleaned['Typ oferty'] == 'Najmu')]
```

```python
data_cleaned['Balkon'] = data_cleaned['Balkon'].replace({'Nie': 0, 'Tak': 1})
data_cleaned['Sprzedaż'] = data_cleaned['Typ oferty'].replace({'Najmu': 0, 'Sprzedaży': 1})
data_cleaned['Najem'] = data_cleaned['Typ oferty'].replace({'Najmu': 1, 'Sprzedaży': 0})
data_cleaned['Rynek pierwotny'] = data_cleaned['Rynek'].replace({'Pierwotny': 1, 'Wtórny': 0})
data_cleaned.drop(columns=['Rynek', 'Typ oferty'], inplace=True)
```

```python
data_cleaned.info()
```

```python
# only temporary, for initial analysis
data_cleaned.to_csv("data_cleaned.csv", index=False)
```
